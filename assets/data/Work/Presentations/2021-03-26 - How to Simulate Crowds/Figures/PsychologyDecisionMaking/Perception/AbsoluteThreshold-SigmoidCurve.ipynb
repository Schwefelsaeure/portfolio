# Plot a sigmoid curve representing the absolute threshold

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
def use_custom_plot_settings(font_weight="normal"):
    font_size_extra_small = 12
    font_size_small = 16
    font_size_medium = 18
    font_size_big = 20

    plt.style.use("default")
    
    plt.rc("font", size=font_size_small, weight=font_weight)
    plt.rc("axes", titlesize=font_size_big, titleweight=font_weight)
    plt.rc("axes", labelsize=font_size_medium, labelweight=font_weight)
    plt.rc("xtick", labelsize=font_size_small)
    plt.rc("ytick", labelsize=font_size_small)
    plt.rc("legend", fontsize=font_size_extra_small)
    plt.rc("figure", titlesize=font_size_big, titleweight=font_weight)

def use_default_plot_settings():
    plt.rcdefaults()
    
use_custom_plot_settings(font_weight="normal")
print(plt.style.available)
```

```python
sigmoid = lambda x: 1 / (1 + np.exp(-x))
x1 = np.linspace(-8, 8, 100)
y1 = sigmoid(x1)

x2 = np.linspace(-8, 8, 11, endpoint=True)
y2 = sigmoid(x2)

fig, ax = plt.subplots()

# Plot function
ax.plot(x1, y1, label="Sigmoid curve")
ax.scatter(x2, y2, s=100, label="Points on curve")

# Plot additional caption
ax.vlines(0, 0, 1, color="gray", linestyles="dashed")
ax.text(1, 0.2, "Absolute threshold\n(50% detection)")
ax.arrow(1, 0.35, -0.85, 0.135)

ax.set_xlabel("Stimulus intensity [arbitrary unit]")
ax.set_ylabel("Detection rate [%]")

ax.set_xticks([-8, 8])
ax.set_xticklabels(["\"Low\"", "\"High\""])

ax.set_yticks([0, 0.5, 1])
ax.set_yticklabels(["0", "50", "100"])

ax.grid(True)
for spine in ["top", "bottom", "left", "right"]:
    ax.spines[spine].set_visible(False)

# ax.legend()

fig.savefig("AbsoluteThreshold-SigmoidCurve.pdf", bbox_inches="tight", transparent=True)
```
